```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\Utkut\OneDrive\Masaüstü\Utku\Coding\Python\Machine Learning\Regression\Data\Ads_CTR_Optimisation.csv')
print(data)
```

```python
import random

n = 10000
adds = 10
total = 0
choosen = []

for n in range(0,n):
    add = random.randrange(adds)
    choosen.append(add)
    reward = data.values[n,add]
    total += reward

plt.hist(choosen)
print(total)
```
